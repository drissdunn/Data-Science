## Data Visualization with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('tested.csv')
df = pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
plt.plot(data['Embarked'], data['Pclass'], label = "Ransom")
```

```python
plt.plot(data.Pclass, data.index, label = "Age")
plt.plot(data.Embarked, data.index, label = "Embarked")

plt.title("Chart 1")
plt.ylabel("index")
plt.legend()
plt.show()
```

```python
plt.plot(df.Species, df.SepalLengthCm, label = "Sepal Length")
plt.plot(df.Species, df.SepalWidthCm, label = "Sepal width")
plt.plot(df.Species, df.PetalLengthCm, label = "Petal Length")
plt.plot(df.Species, df.PetalWidthCm, label = "Petal Width")

plt.title("Chart 2")
plt.ylabel("Index")
plt.legend()
plt.show()
```

```python
plt.plot(df.index, df.SepalWidthCm, label = "SepalWidth", color = "Red" )
plt.ylabel("SepalWidth")
plt.xlabel("Index")
plt.title("Chart 3")
```

```python
plt.scatter(df.index, df.SepalWidthCm, label = "SepalWidth", color = "Orange", linestyle = ':', marker = 's')
plt.style.use("fivethirtyeight")
plt.ylabel("SepalWidth")
plt.xlabel("Index")
plt.title("Chart 4")
```

```python
plt.bar([1, 2, 3, 6, 5, 4, 9, 10, 3], [4, 6, 2, 7, 3, 8, 9, 5, 3], color = 'Brown', label = "List 1")
plt.bar([1, 4, 7, 1, 3, 6, 9, 0, 7], [3, 7, 8, 2, 2, 6,9, 8, 3], color = 'Green', label = "List 2")
plt.legend()
plt.ylabel("Frequency")
plt.xlabel("Item")
plt.title("Chart 5 \n Cool Bar Chart")
```

```python
x = np.linspace(5, 10, 25)
y = x * x + 2

np.array([x]).reshape(1, -1)
```

```python
pyl.plot(x, y, color = 'b')
```

```python
pyl.subplot(1, 2, 1)
pyl.plot(x, y, color = 'r', linestyle = '--', marker = "v")
pyl.subplot(1, 2, 2)
pyl.plot(x, y, color = 'b', linestyle = '--', marker = "s")
```

```python
fig = plt.figure()

axis = fig.add_axes([0.1, 0.2, 0.8, 0.8])

axis.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x, y, color = 'g', linestyle = "--")
    continue
    ax.plot(x, y, color = 'g', linestyle = ":")
```
